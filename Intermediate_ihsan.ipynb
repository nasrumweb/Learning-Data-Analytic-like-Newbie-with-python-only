```python
import matplotlib.colors as mcolors

# Define the green-to-red color map
colors = ['green', 'yellow', 'orange', 'red']
cmap = mcolors.LinearSegmentedColormap.from_list('', colors)

# Convert date column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Calculate total daily transaction volume for each investment type
df2['total_saham_transaction_amount'] = df2['buy_saham_transaction_amount'] - df2['sell_saham_transaction_amount']
df2['total_pasar_uang_transaction_amount'] = df2['buy_pasar_uang_transaction_amount'] - df2['sell_pasar_uang_transaction_amount']
df2['total_pendapatan_tetap_transaction_amount'] = df2['buy_pendapatan_tetap_transaction_amount'] - df2['sell_pendapatan_tetap_transaction_amount']
df2['total_campuran_transaction_amount'] = df2['buy_campuran_transaction_amount'] - df2['sell_campuran_transaction_amount']

# Group data by day and calculate total transaction volume for each investment type
df_by_day = df2.groupby('date').sum()


# Calculate total daily invested amount for each investment type
df2['total_saham_invested_amount'] = df2['saham_invested_amount']
df2['total_pasar_uang_invested_amount'] = df2['pasar_uang_invested_amount']
df2['total_pendapatan_tetap_invested_amount'] = df2['pendapatan_tetap_invested_amount']
df2['total_campuran_invested_amount'] = df2['campuran_invested_amount']

# Group data by day and calculate total invested amount for each investment type
df_by_day = df2.groupby('date').sum()

# Create line plot for total daily invested amount for each investment type
fig, ax = plt.subplots(figsize=(14,6))
ax.plot(df_by_day.index, df_by_day['total_pendapatan_tetap_invested_amount'], marker='o', color=cmap(0.0), label='Pendapatan Tetap')
ax.plot(df_by_day.index, df_by_day['total_pasar_uang_invested_amount'], marker='o', color=cmap(0.33), label='Pasar Uang')
ax.plot(df_by_day.index, df_by_day['total_saham_invested_amount'], marker='o', color=cmap(0.66), label='Saham')
ax.plot(df_by_day.index, df_by_day['total_campuran_invested_amount'], marker='o', color=cmap(1.0), label='Campuran')
ax.set_xlabel('Date')
ax.set_ylabel('Invested Amount')
ax.set_title('Daily Invested Amount by Investment Type')
ax.legend()


# Format the y-axis tick labels as Indonesian Rupiah
def rupiah_format(x, pos):
    return 'Rp{:,.0f}'.format(x)

ax.yaxis.set_major_formatter(plt.FuncFormatter(rupiah_format))

ax.yaxis.set_ticks_position('right')

plt.show()
```

```python
# Drop unnecessary columns from df1
df11 = df1[['user_id', 'user_age', 'user_occupation', 'user_income_range']]
df22 = df2[['user_id', 'date', 'total_saham_invested_amount', 'total_pasar_uang_invested_amount', 'total_pendapatan_tetap_invested_amount', 'total_campuran_invested_amount']]
# Merge df1 and df2 on 'user_id' column
df4 = pd.merge(df11, df22, on='user_id')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df4_cluster = df4.drop(columns=['date'])

# Encode categorical variables in df4_cluster
df4_cluster = pd.get_dummies(df4_cluster, columns=['user_occupation', 'user_income_range'])

# Scale the data using StandardScaler
scaler = StandardScaler()
df4_cluster_scaled = scaler.fit_transform(df4_cluster)

# Determine the optimal number of clusters using the elbow method
def elbow_method(X):
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)
    plt.plot(range(1, 11), wcss)
    plt.title('Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
plt.show()


elbow_method(df4_cluster_scaled)
```

```python
# Based on the elbow method, we choose 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(df4_cluster_scaled)

# Add cluster labels to df4_cluster
df4_cluster['cluster'] = kmeans.labels_

# Map cluster labels to meaningful names
df4_cluster['cluster'] = df4_cluster['cluster'].map({0: 'Conservative', 1: 'Moderate', 2: 'Aggressive'})

# Visualize the clusters
sns.pairplot(df4_cluster, hue='cluster')
plt.show()
```

```python
# Define the balance thresholds for each portfolio level
platinum_threshold = 500000000 # Example threshold, adjust as necessary
gold_threshold = 100000000 # Example threshold, adjust as necessary

# Create the "portfolio_level" variable based on the balance thresholds
df4['portfolio_level'] = pd.cut(df4['total_saham_invested_amount'] + df4['total_pasar_uang_invested_amount'] +
                                 df4['total_pendapatan_tetap_invested_amount'] + df4['total_campuran_invested_amount'],
                                 bins=[-float('inf'), gold_threshold, platinum_threshold, float('inf')],
                                 labels=['Bronze Member', 'Gold Member', 'Platinum Member'])

```

```python
from sklearn.cluster import KMeans
import seaborn as sns

# Extract the relevant columns for clustering
data_for_clustering = df4[['total_saham_invested_amount', 'total_pasar_uang_invested_amount', 'total_pendapatan_tetap_invested_amount', 'total_campuran_invested_amount']]

# Create the k-means model with k=3
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the model to the data
kmeans.fit(data_for_clustering)

# Add the cluster labels to the original dataset
df4['cluster_label'] = kmeans.labels_

# Visualize the clusters
sns.scatterplot(x='total_saham_invested_amount', y='total_pasar_uang_invested_amount', hue='cluster_label', data=df4)

```

```python

# Normalize the total_investment column
df4['total_investment'] = (df4['total_investment'] - df4['total_investment'].mean()) / df4['total_investment'].std()

# Use k-means clustering to create 3 clusters based on the normalized total_investment column
kmeans = KMeans(n_clusters=3, random_state=42).fit(df4[['total_investment']])

# Assign each customer to a cluster based on their total_investment
df4['cluster'] = kmeans.predict(df4[['total_investment']])

# Map the cluster labels to the membership levels
cluster_mapping = {0: 'Bronze Member', 1: 'Gold Member', 2: 'Platinum Member'}
df4['membership_level'] = df4['cluster'].map(cluster_mapping)
```

```python
# Create a bar plot to visualize the total investment amount by membership level
fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(data=df4, x='membership_level', y='total_investment', ax=ax)

# Set the y-axis tick labels to Indonesian Rupiah
ticklabels = ticker.FuncFormatter(lambda x, pos: 'Rp{:,.0f}'.format(x))
ax.yaxis.set_major_formatter(ticklabels)

# Add labels and title to the plot
ax.set_xlabel('Membership Level')
ax.set_ylabel('Total Investment Amount (IDR)')
ax.set_title('Total Investment Amount by Membership Level')

plt.show()
```
